```python
# 1. On pivote
pivot_df = grouped_data.pivot(
    index=['annee', 'arrondissement'],      
    columns='statut_juridique',              
    values=['total_logements', 'total_logements_renove', 'pourcentage_renove']
).reset_index()

# 2. On aplatit les noms des colonnes
pivot_df.columns = [f'{col[0]}_{col[1]}'.lower() if col[1] else col[0] 
                    for col in pivot_df.columns.values]

# --- ASTUCE POUR ÉVITER L'ERREUR ---
# On affiche les colonnes générées pour être SÛR du nom exact
print("Colonnes générées par le pivot :", pivot_df.columns.tolist())

# On définit un dictionnaire de renommage flexible
# Vérifiez bien si c'est 'social' ou 'social (collectif)' ou autre dans votre print ci-dessus
mapping = {
    'total_logements_privé (individuel)': 'logements_prives',
    'total_logements_social': 'logements_sociaux',
    'total_logements_renove_privé (individuel)': 'logements_prives_renoves',
    'total_logements_renove_social': 'logements_sociaux_renoves',
    'pourcentage_renove_privé (individuel)': 'pourcentage_prives_renoves',
    'pourcentage_renove_social': 'pourcentage_sociaux_renoves'
}

# On renomme seulement les colonnes qui existent réellement dans pivot_df
pivot_df = pivot_df.rename(columns=mapping)

# Si une colonne manque (ex: pas de social), on la crée avec des 0 pour éviter que le calcul plante
for col in ['logements_prives', 'logements_sociaux', 'logements_prives_renoves', 'logements_sociaux_renoves']:
    if col not in pivot_df.columns:
        pivot_df[col] = 0

# 4. On remplace les cases vides (NaN) par 0
pivot_df = pivot_df.fillna(0)

# 5. Maintenant le calcul ne plantera plus !
pivot_df['total_logements'] = pivot_df['logements_prives'] + pivot_df['logements_sociaux']
pivot_df['total_logements_renoves'] = pivot_df['logements_prives_renoves'] + pivot_df['logements_sociaux_renoves']

# Pourcentage global (privé + social)
pivot_df['pourcentage_total_renove'] = 0
mask = pivot_df['total_logements'] > 0
pivot_df.loc[mask, 'pourcentage_total_renove'] = ((pivot_df['total_logements_renoves'] / pivot_df['total_logements']) * 100).round(1)

print("\n--- Tableau final réussi ---")
print(pivot_df.head())
```

```python
import pandas as pd
import json
import os

class TableFactory:
    def __init__(self, df_light, config_json_path):
        """
        Au moment de la construction, on reçoit le DataFrame nettoyé
        et le chemin vers la configuration des colonnes.
        """
        self.df = df_light
        with open(config_json_path, 'r', encoding='utf-8') as f:
            self.config = json.load(f)
        
        # Ce dictionnaire stockera les DataFrames générés
        self.generated_tables = {}

    def _generate_single_table(self, table_name):
        """
        MÉTHODE PRIVÉE : Fabrique une table à partir du nom dans le JSON.
        """
        if table_name not in self.config:
            print(f"⚠️ La table '{table_name}' n'est pas dans le fichier config.")
            return None
        
        cols = self.config[table_name]
        print(f"🔨 Génération de {table_name}...")
        return self.df[cols].copy()

    def generate_all(self):
        """
        Boucle sur le JSON pour fabriquer toutes les tables d'un coup.
        """
        for table_name in self.config.keys():
            self.generated_tables[table_name] = self._generate_single_table(table_name)

    def export_to_json(self, output_dir):
        """
        Exporte les tables générées en fichiers JSON pour le Front-end.
        """
        os.makedirs(output_dir, exist_ok=True)
        
        for name, df in self.generated_tables.items():
            path = os.path.join(output_dir, f"{name}.json")
            df.to_json(path, orient='records', force_ascii=False, indent=2)
            print(f"✅ Exporté : {path}")

    def get_json_payload(self):
        """
        Au lieu de sauvegarder des fichiers, cette méthode retourne 
        directement les données au format JSON pour un Contrôleur API.
        """
        payload = {}
        for name, df in self.generated_tables.items():
            # On transforme le DF en dictionnaire Python (format JSON compatible)
            payload[name] = df.to_dict(orient='records')
        return payload
```
